```
Courses = spark.read.format("csv").option("header", "true").load("dbfs:/FileStore/FileStore/Courses-2.csv")
Courses.createOrReplaceTempView("Courses")
Department = spark.read.format("csv").option("header", "true").load("dbfs:/FileStore/FileStore/Department-3.csv")
Department.createOrReplaceTempView("Department")
Schedule = spark.read.format("csv").option("header", "true").load("dbfs:/FileStore/FileStore/Schedule-1.csv")
Schedule.createOrReplaceTempView("Schedule")
Professor = spark.read.format("csv").option("header", "true").load("dbfs:/FileStore/FileStore/Professor-1.csv")
Professor.createOrReplaceTempView("Professor")
```

```
%sql

Select Schedule.PROFESSOR_ID, Courses.ID as Course_id, Courses.NAME as COURSE_NAME, Professor.NAME AS Professor_name, Professor.DEPARTMENT_ID as p_dept_id, Courses.DEPARTMENT_ID as c_dept_id from Schedule 
join Courses on Schedule.COURSE_ID = Courses.ID
join Professor on Professor.ID = Schedule.PROFESSOR_ID
where Professor.DEPARTMENT_ID <> Courses.DEPARTMENT_ID
 
```

```
%sql

select d.Name, p.Salary,p.Name as Prof_name, d_rank from Department d join 
(select *,dense_rank() over(order by salary) as d_rank from Professor) p
on d.ID = p.DEPARTMENT_ID
where d_rank in (1,2,3)
```
